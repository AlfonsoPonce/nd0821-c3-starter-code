```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
data = pd.read_csv(str(Path('data/census.csv')))
data
```

```python
data.info()
```

*MISSING VALUES COUNT*

```python
data.isna().sum()
```

*IDENTIFYING COLUMN TYPES*

```python
data.dtypes
```

```python
target_variable = 'salary'
print(f"TARGET VARIABLE -> {target_variable}")

integer_variable_list = [c.strip() for c in data.columns if data[c].dtype == 'int64']
print(f"INTEGER VARIABLES-> {integer_variable_list}")

categorical_variable_list = [c.strip() for c in data.columns if data[c].dtype == 'object' and c.strip() != target_variable]
print(f"CATEGORICAL VARIABLES -> {categorical_variable_list}")
```

*REMOVING TRAILING WHITESPACE*

```python
data.columns = [c.strip() for c in data.columns]
data['native-country'].unique().tolist()
```

*SAVING DATA*

```python
data.to_csv('cleaned_data.csv')
```
